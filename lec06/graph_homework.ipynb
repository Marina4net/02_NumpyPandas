```python
# подгружаем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
power = pd.read_csv('power.csv') 
power.sample(8)
```

```python
# количество данных по годам - Правда ли что чем больше год, тем больше данных
year_group = power.groupby('year').count().sort_values('quantity', ascending = False).reset_index()
plt.barh(year_group['year'], year_group['country'])

plt.xlabel('Quantity')
plt.ylabel('Year')
plt.title('Data per year')
plt.grid()

#вывод - однозначно - с годами собираемость данных растет
```

```python
#потребление по годам стран прибалтики
countryes = ['Lithuania', 'Latvia', 'Estonia']
Lithuania = power[power['country'] == 'Lithuania'].groupby(['year']).sum().quantity
Latvia = power[power['country'] == 'Latvia'].groupby(['year']).sum().quantity
Estonia = power[power['country'] == 'Estonia'].groupby(['year']).sum().quantity

Lithuania.plot(label = u'Литва')
Latvia.plot(label = u'Латвия')
Estonia.plot(label = u'Эстония')
plt.legend()
#Эстония оказывается страна с потрясающим потенциалом по потреблению. Исследуем это дело
```

```python
Estonia2000 = power[(power['country'] == 'Estonia') & (power['year'] == 2000)].groupby(['category']).sum()
Estonia2004 = power[(power['country'] == 'Estonia') & (power['year'] == 2004)].groupby(['category']).sum()

Estonia2000.quantity.plot(kind='bar', rot=45, label = u'2000')
Estonia2004.quantity.plot(kind='bar', rot=45, color='m', label = u'2004')
plt.legend()
# катастрофический рост 58 категории. наверное там зашита промышленность, которая то поднималась то умирала. 
```

```python
Estonia2000[Estonia2000.index != 58].quantity.plot(kind='box')
#Если построить ящик с усами то мы видим сплошные выколотые точки, даже если выкинули 58 категорию. Может нам поможем seaborn
```

```python
#но перед этим посмотрим на эстонский ящик 2004 года
Estonia2004.quantity.plot(kind='box')
```
