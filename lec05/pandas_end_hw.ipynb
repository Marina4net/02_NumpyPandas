# Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
```

```python
raiting = pd.read_csv('ratings.csv')
raiting.head()
```

```python
rg = raiting.groupby('userId').agg(['max', 'min'])['timestamp']
rg['lifetime'] = (rg['max'] - rg['min'])/3600/24 #время жизни пользователя в днях 

obj = raiting.groupby('userId').count()
obj = obj[(obj['rating'] > 100)] #пользователи с больше чем 100 комментариями

m = obj.join(rg, how='left')
ave = m.lifetime.mean()
print(f'среднее время жизни пользователя, поставившего больше 100 оценок - {ave} суток')
```
